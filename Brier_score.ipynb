```python
import pickle
from pathlib import Path
from sklearn.metrics import brier_score_loss
```

```python
preds_pickle_file_path = Path("13-5_fold_cross_validation-copi_val-separated=False--2.pkl")
```

```python
preds_pickle_file_path.exists()
```

```python
def load_pickle(filepath):
    f = open(filepath,'rb')
    file = pickle.load(f)
    f.close()
    return file
```

```python
copi_val = load_pickle(preds_pickle_file_path)
```

```python
copi_val
```

```python
for i, Class in enumerate(['Corticotropic', 'Silent Corticotropic', 'Luteinizing Hormone', 'Follicle Stimulating Hormone']):
    y_preds_raw = [] # list of the predicted percentages
    y_true = [] # list of True and False
    for key, value in copi_val.items():
        y_preds_raw.append(value[1][i]/value[0])
        y_true.append(value[2][i])
    
    b = brier_score_loss(y_true=y_true, y_prob=y_preds_raw)
    
    print(f'{Class}: {b}')  
```
